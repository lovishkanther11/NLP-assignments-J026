```
import pandas as pd
import gzip
```

```
df = pd.read_json("http://snap.stanford.edu/data/amazon/productGraph/categoryFiles/reviews_Office_Products_5.json.gz" , lines=True)
```

```
df.head()
```

```
def wc(str):
    count = dict()
    word = str.split()

    for l in word:
        if l in count:
            count[l] += 1
        else:
            count[l] = 1

    return count
```

```
df['reviewText_wc'] = df['reviewText'].apply(wc)
df['summary_wc'] = df['summary'].apply(wc)
```

```
df.head()
```

```
vowels = ['a','e','i','o','u']
```

```
import re
def split(str):
  str = re.sub('[^a-zA-Z]',' ', str)
  return list(str)
```

```
def let(text):
  text = text.lower()
  dict = {"vowels":0,
          "consonants":0}
  for word in text:
    words = split(word)
    for p in words:
      if p in vowels:
        dict["vowels"] += 1
      else:
         dict["consonants"] += 1
  return dict
```

```
df['reviewText_vowel_consonant'] = df['reviewText'].apply(lambda a:let(a))
df['summary_vowel_consonant'] = df['summary'].apply(lambda b:let(b))
```

```
df.head()
```

```
def wordfind(word):
  word=re.sub('[^a-zA-Z]',' ',word)
  list_b=re.findall(r'\b[bB]\w+',word)
  list_g=re.findall(r'\w+[Gg]\b',word)
  return (list(set().union(list_b,list_g)))
```

```
def beginG(word):
  word=re.sub('[^a-zA-Z]',' ',word)
  words=re.findall(r'\b[bB]\w+[Gg]\b',word)
  return words
```

```
df['reviewText_b_or_g']=df['reviewText'].apply(wordfind)
df['summary_b_or_g']=df['summary'].apply(wordfind)
df['reviewText_b_and_g']=df['reviewText'].apply(beginG)
df['summary_b_and_g']=df['summary'].apply(beginG)
```
